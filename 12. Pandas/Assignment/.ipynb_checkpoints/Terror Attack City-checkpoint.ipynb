## Terror Attack City

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period.

#### Problem Statement:
The Most Dangerous city in Jammu and Kashmir and the terrorist group which is most active in that city?<br>
<br>
Print count of number of attacks in that city as integer value.<br>
<br>
__Note:__ Ignoring the Unknown Terrorist Group. Here Dangerous related with number of terrorist attacks.<br>
<br>
#### Output Format:
City  NumberOfAttack   Group

```python
# Importing the libraries
import numpy as np
import pandas as pd 
```

```python
dataset = pd.read_csv('terrorismData.csv')
dataset
```

```python
# making a copy of the dataset
df = dataset.copy()
```

```python
# Extracting the part of dataset where the State is 'Jammu and Kashmir'
df1 = df[df['State'] == 'Jammu and Kashmir']
df1
```

```python
# Finding out the description of the City column
df2 = df1.City.describe()
df2
```

```python
count = df2.freq
count
```

```python
# Extracting the part of the dataset from df1 where the Terrorist Group is known
df3 = df1[df1.Group != 'Unknown']
df3
```

```python
# Extracting the maximum no.of time a City has been attacked
city = df3.City.describe().top
city
```

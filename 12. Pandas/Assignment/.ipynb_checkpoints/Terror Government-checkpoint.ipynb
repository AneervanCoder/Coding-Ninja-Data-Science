## Terror Government

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period.

#### Problem Statement:
There was formation of new government in India on 26 May 2014. So current government's span is from 26th May 2014 to current. Find out two things from this period-<br>
<br>
1. Total number of attacks done in this period in India. Find this count as integer
2. Which Terrorist group was most active in this period in India. Most active means, group which has done maximum number of attacks.
3. Ignore the Unknown group

#### Output Format:
TotalAttacks MostActive TerroristGroup

```python
# Importing the libraries
import numpy as np
import pandas as pd 
```

```python
terror_data = pd.read_csv('terrorismData.csv')
```

```python
# making the copy of the dataset
df =  terror_data.copy()
df
```

```python
# Extracting the part of the dataset of 26th May 2014, India
df1 = df[(df['Year'] == 2014) & (df['Month'] == 5) & (df['Day'] == 26) & (df['Country'] == 'India')]
df1
```

```python
# a = df1[df1['Group'] != 'Unknown']
# a
```

```python
a = df1.describe()
a
```

```python
count = a.loc['count', 'Year']
count = int(count)
count
```

```python
data = {'Total Attacks': count, 'Most Active': 'Unknown', 'Terrorist Group': 'Unknown'}
answer = pd.DataFrame(data, index=[0])
answer
```

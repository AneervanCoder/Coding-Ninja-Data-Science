## Terror Attack

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period.

#### Problem Statement:
Find out the Country with Highest Number of Terror Attack and in  which year the most number of terrorist attack happened in that Country?<br>
<br>
Print count of terror attacks as integer value.<br>

#### Output Format:
Country  NumberOfAttack   Year

```python
# Importing the libraries
import numpy as np
import pandas as pd 
```

```python
dataset = pd.read_csv('terrorismData.csv')
```

```python
# making the copy of the dataset
df = dataset.copy()
df
```

```python
# Extracting the Country from the dataset in which there is highest number of Terror Attack
df1 = df.Country.describe().top 
df1
```

```python
country = df1
country
```

```python
# Extracting the part of the dataset for that Country only
df_iraq = df[df['Country'] == 'Iraq']
df_iraq
```

```python
# Creating the dataframe of the year description
year_df = pd.DataFrame({'Year Description': df_iraq.Year.describe()})
year_df
```

```python
# Extracting the count from the dataframe
count = year_df.loc['count', 'Year Description']
count = int(count)
count
```

```python
# exatracting the year from the dataframe
max_year = year_df.loc['max', 'Year Description']
max_year = int(max_year)
max_year
```

```python
print('Country \t NumberOfAttack \t Year')
print(country,'\t\t\t', count,'\t\t', max_year)
```

```python
# Creating the above ouput as a dataframe
data = {'Country':country, 'Number of Attack':count, 'Year':max_year}
answer = pd.DataFrame(data, index=[0])
answer
```

## Terror Deadliest Attack

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period.

#### Problem Statement:
Most Deadliest attack in a history of HumanKind?<br>
<br>
Print count of killed people as integer value.<br>
<br>
__Note:__ Here Deadliest attack means, in which the most number of people killed

#### Output Format:
NumberOfPeopleKilled Country TerroristGroup

```python
# Importing the libraries
import numpy as np
import pandas as pd 
```

```python
dataset = pd.read_csv('terrorismData.csv')
```

```python
#making the copy of the dataset
df = dataset.copy()
df
```

```python
# the row which has the maximum Killed value
df_killed = df[df.Killed == df.Killed.max()]
df_killed
```

```python
people_killed = df_killed.loc[133518,'Killed']
people_killed = int(people_killed)
people_killed
```

```python
country = df_killed.loc[133518,'Country']
country
```

```python
group = df_killed.loc[133518, 'Group']
group
```

```python
print('NumberOfPeopleKilled \t Country \t TerroristGroup')
print(people_killed,'\t\t\t', country, '\t', group)
```

```python
data = {'Number of People Killed':people_killed, 'Country':country, 'Terrorist Group':group}
answer = pd.DataFrame(data, index=[0])
answer
```

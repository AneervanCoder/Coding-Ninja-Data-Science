### Handling String

```python
# string based data
df['Gender'] = "Female"
df.iloc[0:10, 6] = "Male"
df.tail()
```

```python
def f(s):
    if(s == "Male"):
       return 0
    else:
       return 1
```

* __DataFrame.apply(_func, axis=0, raw=False, result_type=None, args=(), **kwargs_)__: _Apply a function along an axis of the DataFrame_

```python
df['sex'] = df.Gender.apply(f)
del df['Gender']
```

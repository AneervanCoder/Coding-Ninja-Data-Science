## Plot Job_java

It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon Job site.

#### Problem Statement:
Plot the scatter graph between year vs No.of jobs opening related to java.<br>
<br>
Print the year and number of Jobs opening in Java Profile.<br>
<br>
__Note:__ Use the keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.

#### Ouput Format:
year1 JobOpening1<br>
year2 JobOpening2<br>
....<br>
....<br>
....<br>

```python
# importing the libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
amazon_jobs = pd.read_csv('amazon_jobs_dataset.csv')
# making the copy of the dataset
df = amazon_jobs.copy()
df
```

```python
# droping the column name Unnamed:0 as it is not required
df1 = df.drop('Unnamed: 0', axis=1)
df1
```

```python
# extracting the part of the dataset which has a basic qualification as 'Java'
# boolean series returned with False at place with NaN
bool_series = df1['BASIC QUALIFICATIONS'].str.contains('Java', na=False, case=False)
df2 = df1[bool_series]
df2
```

```python
#Extracting the posting date and the qualification dataset
mycolumns = ['Posting_date', 'BASIC QUALIFICATIONS']
df3 = df2.loc[:, mycolumns]
df3
```

```python
df3.columns
```

```python
# Converting the date format to yyyy-mm-dd
df3['Posting_date'] = pd.to_datetime(df3.Posting_date)
df3['Year'] = df3['Posting_date'].dt.year  # Extracting the year from the datetime object
df3.sort_values(by='Year', ascending=True, inplace=True)
del df3["Posting_date"]
```

```python
del df3["BASIC QUALIFICATIONS"]
```

```python
df3.dropna(inplace=True)
df3
```

```python
java_jobs = pd.DataFrame(df3.Year.value_counts(ascending=True))
java_jobs
```

```python
# changing the column name
java_jobs.columns = ['Jobs']
java_jobs
```

```python
# Plotting the scatter graph
X = java_jobs.index
y = java_jobs.Jobs
plt.scatter(X, y, c='r')
plt.xlabel('Years')
plt.ylabel('No.of Java Jobs')
plt.title('Year Vs No.of Java Jobs')
plt.show()
```

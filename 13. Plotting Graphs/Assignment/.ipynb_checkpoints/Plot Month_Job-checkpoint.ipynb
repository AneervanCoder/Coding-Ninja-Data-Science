## Plot Month_Job

It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon Job site

#### Problem Statement:
Plot the Bar graph between Month vs Job Openings.<br>
<br>
Print the year and the number of job posting as integer value.<br>
<br>
#### Ouput Format:
month_name1 job_posting1<br>
month_name2 job_posting2<br>
....<br>
....<br>
....<br>

```python
# Imprting the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset and making its copy
amazon_jobs = pd.read_csv('amazon_jobs_dataset.csv')
df = amazon_jobs.copy()
df = pd.DataFrame(amazon_jobs)
df
```

```python
# extracting the Posting_date column
jobs_date = df[['Posting_date']]
jobs_date.tail()
```

```python
# Converting the date format to yyyy-mm-dd
jobs_date['Posting_date'] = pd.to_datetime(jobs_date.Posting_date)
jobs_date['Month'] = jobs_date['Posting_date'].dt.month_name() # Extracting the month name from the datetime object
jobs_date.tail()
```

```python
# Extracting the month column from the dataframe and sorting it
df_m= pd.DataFrame(jobs_date.loc[:, 'Month'])
df_m.tail()
```

```python
# counting how many times a particular month has occurred in the dataframe
df_m = pd.DataFrame(df_m.Month.value_counts())
df_m
```

```python
# X and y axis
X = df_m.index
y = df_m['Month']
```

```python
# Ploting a bar groph
plt.bar(X, y)
plt.title('Amazon Jobs Opening in a Month')
plt.xlabel('Months')
plt.ylabel('No.of Jobs Openings')
plt.xticks(rotation = 40)
plt.show()
```

## Plot Job Year

It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon Job site

#### Problem Statement:
Plot the line graph between no.of job postings with respect to year.<br>
<br>
Print the year and the number of job posting as integer value.<br>
<br>
__Note:__ Year should be in ascending order

#### Ouput Format:
year1 job_posting1<br>
year2 job_posting2<br>
....<br>
....<br>
....<br>

```python
# Imprting the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Importing the dataset and making its copy
amazon_jobs = pd.read_csv('amazon_jobs_dataset.csv')
df = amazon_jobs.copy()
df = pd.DataFrame(amazon_jobs)
df
```

```python
# extracting the Posting_date column
jobs_date = df[['Posting_date']]
jobs_date.tail()
```

```python
# Converting the date format to yyyy-mm-dd
jobs_date['Posting_date'] = pd.to_datetime(jobs_date.Posting_date)
jobs_date['Year'] = jobs_date['Posting_date'].dt.year  # Extracting the year from the datetime object
jobs_date.head()
```

```python
# Sorting the date and year coloumn
jobs_date.sort_values(by='Posting_date')
```

```python
# removing the any NaN values
jobs_date.dropna()
jobs_date.tail()
```

```python
# Extracting the year column from the dataframe and sorting it
df_y= pd.DataFrame(jobs_date.loc[:, 'Year'])
df_y.sort_values('Year', inplace=True)
df_y.tail()
```

```python
# counting how many times a particular year has occurred in the dataframe
df_y = pd.DataFrame(df_y.Year.value_counts())
df_y
```

```python
# X and y axis
X = df_y.index
y = df_y['Year']
```

```python
# Ploting a line groph
plt.plot(X, y)
plt.title('Amazon Jobs Posting in a Year')
plt.xlabel('Years')
plt.ylabel('No.of Jobs Posting')
plt.show()
```

### Customizing Graph

```python
# color, marker, linewidth, labels for axis, datasets and the title, axis grid, text
# x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
#plt.plot(x,y, color = 'black', marker = "p")
plt.plot(x,y, color = 'black', linewidth = 5, label="x^3")

plt.plot(x,y2, color = 'red', linewidth = 5, label="x^2")

# plt.axis([0, 10, 0, 200])

plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2, 80, "Text", fontsize=12)

plt.legend()
plt.title("Matplotlib demo")
plt.show()
```

### Bubble Chart

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
```

```python
plt.scatter(year, salary, s=population, c='r')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s=population, c='r', alpha=0.1)
plt.title("Bubble Chart")
plt.show()
```

```python
c = np.arange(len(year))
c[1] = c[0]
```

```python
plt.scatter(year, salary, s=population, c=c, alpha=0.5)
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s=population, c=c, alpha=0.5, edgecolor='black')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s=population, c=c, alpha=0.5, edgecolor='black', marker="^")
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s=population, c=c, alpha=0.5, edgecolor='black', marker="^")
plt.title("Bubble Chart")
plt.text(year[0] + 0.1, salary[0] + 0.1, population[0])
plt.text(year[4] + 0.1, salary[4] + 0.1, population[4])
plt.show()
```

### Bar Graphs

### Pie Charts

```python
# pie graphs
labels = ["A","B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["blue", "purple", "red", "pink"]
explode= [0.1, 0.2, 0, 0]
plt.title("Split among classes")
plt.pie(sizes, colors=colors, explode = explode, labels=labels, autopct="%.2f%%", startangle=100)
plt.axis("equal")
plt.show()
```

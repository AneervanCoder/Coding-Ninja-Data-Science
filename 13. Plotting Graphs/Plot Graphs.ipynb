## Plot Graphs

Plot the two graphs on the same plot. The graph between the two given integer array X an Y and the graph between two integer array X and Y where X should hold 20 even points starting from 0 and y should be equal to X where y[i] = x[i]. Plot should have x label and y label and the legend to differentiate between the two graphs. 

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = [0,1,2,3,4,5,6,7,9,10]
y1 = [1,2,5,10,17,26,37,50,82,101]
```

```python
x2 = np.arange(0, 20)
y2 = x2
```

```python
plt.plot(x1,y1)
plt.title("Graph 1")
plt.xlabel("X1")
plt.ylabel("Y1")
plt.grid()
plt.show()
```

```python
plt.plot(x2,y2)
plt.title("Graph 2")
plt.xlabel("X2")
plt.ylabel("Y2")
plt.grid()
plt.show()
```

```python
plt.plot(x1, y1, color='blue', label='Graph 1')
plt.plot(x2, y2, color='red', label='Graph 2')

plt.title("Graph 1 & Graph 2")

plt.xlabel('X - Values')
plt.ylabel('Y - Values')

plt.legend()

plt.show()
```

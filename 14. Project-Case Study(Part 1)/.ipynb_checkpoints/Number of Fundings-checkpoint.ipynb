## Number of Fundings

#### Problem Statement:
Check the trend of investements  over the years. To check the trend, find-<br>
Total number of fundings done in each year.<br>
Plot a line graph between year and number of fundings. Take year on x-axis and number of fundings on y-axis.<br>
Print year-wise total number of fundings also. Print years in ascending order.<br>

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
startup = pd.read_csv('startup_funding.csv')
# making the copy of the dataframe
df = startup.copy()
df
```

```python
#Extracting the Date, AmountInUSD columns
mycolumns = ['Date', 'AmountInUSD']
df2 = df.loc[:, mycolumns]
df2
```

```python
# Handled the error in the Date column
df2.replace(to_replace='22/01//2015', value='22/01/2015', inplace=True)
```

```python
# Converting the date format to yyyy-mm-dd
df2['Date'] = pd.to_datetime(df2.Date)
df2['Year'] = df2['Date'].dt.year  # Extracting the year from the datetime object
df2.tail()
```

```python
# Sorting the above coloumn
df2.sort_values(by='Date', ascending=True, inplace=True)
```

```python
df2
```

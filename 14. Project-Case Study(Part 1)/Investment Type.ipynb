## Investment Type

#### Problem Statement:
There are 4 different type of investments. Find out percentage of amount funded for each investment type.<br>
<br>
Plot a pie chart and visualise.<br>
<br>
Print the investment type and percentage of amount funded with 2 decimal places after rounding off.<br>

__Note:__<br>
Correct spelling of investment types are-"Private Equity","Seed Funding","Debt Funding",and "Crowd Funding". Keep an eye for any spelling mistake. You can find this by printing unique values from this column.<br>
<br>
Print the investment type in descending order with respect to the percentage of the amount funded.<br>
<br>
#### Output Format:
investmentType1 percent1<br>
investmentType2 percent2<br>
investmentType3 percent3<br>
...<br>
...<br>

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
startup = pd.read_csv('startup_funding.csv')
# making the copy of the dataframe
df = startup.copy()
df
```

```python
#Extracting the columns startup name and CityLocation
mycolumns = ['InvestmentType', 'AmountInUSD']
df2 = df.loc[:, mycolumns]
df2
```

```python
# modifying the the column AmountInUSD
def modified(amount):
    return int(amount.replace(',',''))
```

```python
# filling up the NaN values of the columns
df2['InvestmentType'].fillna('', inplace=True)
df2['AmountInUSD'].fillna('0', inplace=True)
df2['AmountInUSD'] = df2['AmountInUSD'].apply(modified)
df2 = df2[df2['InvestmentType'] != ""]
df2
```

```python
#replacing the wrong written investment type
df2['InvestmentType'].replace("SeedFunding", "Seed Funding", inplace = True)
df2['InvestmentType'].replace("PrivateEquity", "Private Equity", inplace = True)
df2['InvestmentType'].replace("Crowd funding", "Crowd Funding", inplace = True)
```

```python
# chechking if the wrong investment type are still present or not
df2['InvestmentType'].isin(['Crowd funding']).any()
```

```python
df2.head()
```

```python
df3 = df2.groupby(by='InvestmentType').sum()
df3
```

```python
# calculating the total amount invested by the investment type
s = df3['AmountInUSD'].sum()
s
```

```python
# calculating the percentage and sorting it in descending order
df4 = pd.DataFrame((df3['AmountInUSD'] / s)*100)
df4.sort_values(by='AmountInUSD', ascending=False, inplace=True)
df4 = pd.DataFrame(df4['AmountInUSD'].round(2))
df4
```

```python
# visualizing the above using pie chart
counts = df4['AmountInUSD']
labels = df4.index
plt.axis("equal")
plt.pie(counts, labels=labels)
plt.show()
```

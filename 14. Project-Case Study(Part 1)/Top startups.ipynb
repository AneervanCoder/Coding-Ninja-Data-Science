## Top startups

#### Problem Statement:
Find top 5 startups with most amount of total funding.<br>
<br>
Print the startup name in descending order with respect to amount of funding.<br>

__Note:__<br>
Ola, Flipkart, Oyo, Paytm are important startups, so correct their names. There are many errors in startup names, ignore correcting all, just handle important ones.
<br>
#### Output Format:
startup1<br>
startup2<br>
startup3<br>
...<br>
...<br>

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
startup = pd.read_csv('startup_funding.csv')
# making the copy of the dataframe
df = startup.copy()
df
```

```python
#Extracting the columns startup name and CityLocation
mycolumns = ['StartupName', 'AmountInUSD']
df2 = df.loc[:, mycolumns]
df2
```

```python
# modifying the the column AmountInUSD
def modified(amount):
    return int(amount.replace(',',''))
```

```python
# filling up the NaN values of the columns
df2['StartupName'].fillna('', inplace=True)
df2['AmountInUSD'].fillna('0', inplace=True)
df2['AmountInUSD'] = df2['AmountInUSD'].apply(modified)
df2
```

```python
#correcting the important startup nmaes
df2["StartupName"].replace("Flipkart.com","Flipkart",inplace = True)
df2["StartupName"].replace("Ola Cabs","Ola",inplace = True)
df2["StartupName"].replace("Olacabs","Ola",inplace = True)
df2["StartupName"].replace("Oyorooms","Oyo",inplace = True)
df2["StartupName"].replace("OyoRooms","Oyo",inplace = True)
df2["StartupName"].replace("OYO Rooms","Oyo",inplace = True)
df2["StartupName"].replace("Oyo Rooms","Oyo",inplace = True)
df2["StartupName"].replace("Paytm Marketplace","Paytm",inplace = True)

df2 = df2[df2["StartupName"] != ""]
df2.head()
```

```python
# checking if the wrong startup name are still present or not
df2['StartupName'].isin(['Olacabs']).any()
```

```python
df3 = df2.groupby(by='StartupName').sum().sort_values(by='AmountInUSD', ascending=False)
df3 = df3.head(5)
```

```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
# loading the dataset
dataset = pd.read_csv('wine_data.csv')
dataset.head()
```

```python
# giving the respective column names to the dataset
columnNames = ['Alcohol' , 'Malic acid' , 'Ash' , 'Alcalinity of ash' , 'Magnesium' ,'Total phenols' , 'Flavanoids','Nonflavanoid phenols' ,'Proanthocyanins','Color intensity' ,'Hue' ,'OD280/OD315 of diluted wines' ,'Proline'     ]

wine = pd.read_csv('wine_data.csv', names=columnNames)
wine.head()
```

## Scaler Method

```python
from sklearn.preprocessing import scale
wine_scaled_1 = scale(wine) # returns numpy array
wine_scaled_1_df = pd.DataFrame(wine_scaled_1, columns=columnNames) # converted numpy array to Dataframe
wine_scaled_1_df
```

## Standard Scaler Method

```python
from sklearn.preprocessing import StandardScaler
wine_scaled_2 = StandardScaler() # returns numpy array
a = wine_scaled_2.fit_transform(wine)
```

```python
# Converting it into Dataframe
wine_scaled_2_df = pd.DataFrame(a, columns=columnNames)
wine_scaled_2_df
```

## Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
wine_scaled_3 = MinMaxScaler() # returns numpy array
b = wine_scaled_3.fit_transform(wine)
```

```python
# Converting it into Dataframe
wine_scaled_3_df = pd.DataFrame(b, columns=columnNames)
wine_scaled_3_df
```
